14. How many people have a credit card that expires in 2020?


```python
def fun():
  cnt=0
  for date in data['CC Exp Date']:
    if date.split('/')[1]=='20':
      cnt=cnt+1
  print(cnt)

fun()
```

```python
len(data[data['CC Exp Date'].apply(lambda x:x[3:]=='20')])
```
